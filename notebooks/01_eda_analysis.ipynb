```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"
column_names = [
    'checking_status', 'duration', 'credit_history', 'purpose', 'credit_amount',
    'savings_status', 'employment', 'installment_commitment', 'personal_status',
    'other_parties', 'residence_since', 'property_magnitude', 'age',
    'other_payment_plans', 'housing', 'existing_credits', 'job',
    'num_dependents', 'own_telephone', 'foreign_worker', 'class'
]  # Lista completa de nombres

df = pd.read_csv(url, delimiter=' ', names=column_names, header=None)

print("="*70)
print("DIAGNÓSTICO DE DATOS CARGADOS")
print("="*70)

# 1. Verificar dimensiones
print(f"1. Dimensiones del DataFrame: {df.shape}")
print(f"   - Filas: {df.shape[0]}")
print(f"   - Columnas: {df.shape[1]}")

# 2. Verificar nombres de columnas
print(f"\n2. Nombres de columnas:")
print(f"   - Total columnas: {len(df.columns)}")
print(f"   - Columnas: {list(df.columns)}")

# 3. Verificar específicamente la columna 'class'
print(f"\n3. ¿Existe la columna 'class'?")
print(f"   - 'class' in df.columns: {'class' in df.columns}")
print(f"   - df.columns[-1] (última columna): {df.columns[-1]}")

# 4. Mostrar primeras filas
print(f"\n4. Primeras 5 filas del DataFrame:")
print(df.head())

# 5. Mostrar últimas filas
print(f"\n5. Últimas 5 filas del DataFrame:")
print(df.tail())

# 6. Verificar valores de la última columna (que debería ser 'class')
print(f"\n6. Última columna (índice 20):")
print(f"   - Nombre: {df.columns[20]}")
print(f"   - Valores únicos: {df.iloc[:, 20].unique()}")
print(f"   - Conteo: {df.iloc[:, 20].value_counts()}")

# 7. Verificar si hay espacios en los nombres
print(f"\n7. Nombres de columnas con espacios/tabs:")
for col in df.columns:
    if ' ' in col or '\t' in col:
        print(f"   - '{col}' tiene espacios")
```

```python
print("Información del dataset:")
print(df.info())
print("\nEstadísticas descriptivas:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='class', data=df)
plt.title('Distribución de Riesgo Crediticio')
plt.xlabel('Clase (1=Buena, 2=Mala)')
plt.ylabel('Frecuencia')
plt.show()
```
